```python
# pip install openpyxl

import pandas as pd

df = pd.read_excel('UpdatedEnglishTenseUniqueDataset.xlsx')

df
```

```python
text = "allah is the greatest"
for sentence in df['Sentence']:
    text += "\n" + sentence
```

```python
print(text)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in text.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
# from tensorflow.keras.utils import to_categorical
# y = to_categorical(y,num_classes=10802)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(10802, 100, input_length=26))
model.add(LSTM(150))
model.add(Dense(10802, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
# from tensorflow.keras.callbacks import EarlyStopping

# early_stop = EarlyStopping(
#     monitor='val_loss',     # watch validation loss
#     patience=2,             # wait 3 epochs before stopping
#     restore_best_weights=True
# )
```

```python
model.fit(X, y, epochs=18, batch_size=32)
```

```python
import pickle

model.save("next_word_model.keras")

with open("tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)

with open("seq_len.pkl", "wb") as f:
    pickle.dump(max_len, f)
```

```python
import numpy as np
def predict_next_word(seed_text, tokenizer, model, max_seq_len):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
    
    predicted = np.argmax(model.predict(token_list), axis=-1)
    
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            return word
```

```python
print(predict_next_word("AI in wildlife conservation will have", tokenizer, model, max_len))
```

```python
inp = "be kind"
for i in range (10):
    pred = predict_next_word(inp, tokenizer, model, max_len)
    inp += f' {str(pred)}'
    print(inp)
```

```python
predict_next_word('if we do exercise daily', tokenizer, model, max_len)
```
